# 2023

```python
a23 = Dep_yearly(2023)
a23.head(3)
```

This graphs the annual salary spend by department by a given year. Here is the data for 2023.

```python
a_dep(a23)
```

This will display all employees whom have averages 66+ hours a week.
<br>
The weekly hours worked are in the far right column. 
* Determine the hourly rate (column Hr_Rate):
* Hr_Rate = Regular_Rate / 2080
* Calculate the overtime rate (column Ot_Rate):
* Ot_Rate = Hr_Rate * 1.5   
* Calculate the number of overtime hours:
* Overtime_Hours = Overtime_Rate / Ot_Rate
* Distribute evenly over 52 weeks:
* Weekly_Overtime_Hours = Overtime_Hours 
* Calculate the weekly average:
* Weekly_Average = Weekly_Overtime_Hours + 40
<br>    
This is not accounting for vacations off time ect. So we would expect some weeks have higher and lower hours.
